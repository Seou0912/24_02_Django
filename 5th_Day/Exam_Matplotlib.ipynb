```python
x = np.arange(0,12,1)
y = np.sin(x)
y2 =np.cos(x)

plt.plot(x,y, label="sin", marker="o")
plt.plot(x,y2, label="cos", marker="x")
plt.legend()
# plt.legend(loc=(1,1)) # 범례의 위치 지정

plt.title("sin graph") # 그래프의 제목
plt.xlabel("Time") # 그래프의 x라벨
plt.ylabel("Amplitude") # 그래프의 y라벨
plt.grid() # 그래프 내 그리드(격자) 생성

# limit 값
plt.xlim(0, np.pi)
plt.ylim(-1, 1)# 위의 사이트에서 가져 온 코드

species = ('Adelie', 'Chinstrap', 'Gentoo')
sex_counts = {
    'Male': np.array([73, 34, 61]),
    'Female': np.array([73, 34, 58]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(3)

for sex, sex_count in sex_counts.items():
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    bottom += sex_count

    ax.bar_label(p, label_type='center')

ax.set_title('Number of penguins by sex')
ax.legend()

plt.show()
```

```python
# 위의 사이트에서 가져 온 코드

species = ('Adelie', 'Chinstrap', 'Gentoo')
sex_counts = {
    'Male': np.array([73, 34, 61]),
    'Female': np.array([73, 34, 58]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(3)

for sex, sex_count in sex_counts.items():
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    bottom += sex_count

    ax.bar_label(p, label_type='center')

ax.set_title('Number of penguins by sex')
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. 영화 데이터를 활용하여 x 축은 Movie, y축은 Rating인 막대 그래프를 만듭니다.
movies = ['Movie A', 'Movie B', 'Movie C', 'Movie D', 'Movie E']
ratings = [4.5, 4.3, 4.8, 4.2, 4.6]

plt.bar(movies, ratings)

# 2. 막대 그래프에 제시된 세부 사항을 적용합니다.
plt.title('Movie Night Top5')
plt.xlabel('Movie', rotation=90)
plt.ylabel('Rating')

plt.show()
```

```python
# 3. 개봉 연도별 평점 변화 추이를 선 그래프로 그립니다.
years = [2000, 2005, 2010, 2015, 2020]
ratings = [8.5, 8.7, 9.1, 9.0, 8.8]

plt.plot(years, ratings)
```

```python
#4. 그래프에 제시된 세부 사항을 적용합니다.
plt.title('Rating Trend by Release Year')
plt.xlabel('Year')
plt.ylabel('Rating')
plt.xticks([2005, 2010, 2015, 2020])
plt.yticks(range(7, 11))
plt.ylim(7, 10)
plt.grid(True)
plt.plot(years, ratings, marker='o')

plt.show()
```

```python
#5. 평점이 9점 이상인 영화의 비율을 확인할 수 있는 원 그래프를 그립니다.
labels = ['9점 이상', '9점 미만']
sizes = [70, 30]

plt.pie(sizes, labels=labels, autopct='%.1f%%')

plt.title("Rating Distribution")
plt.axis('equal')

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')

# 데이터
labels = ['9점 이상', '9점 미만']
sizes = [90.5, 9.5]  # 예시로 임의의 비율 데이터 사용

# 그래프 그리기
colors = ['#FFA500', '#808080']

plt.pie(sizes, labels=labels, colors=colors, autopct='%.1f%%', startangle=90)
plt.legend(loc='center right')

plt.axis('equal')  # 원형 그래프를 그리기 위해 가로세로 비율을 동일하게 설정

plt.show()
```
