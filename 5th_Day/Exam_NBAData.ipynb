```python
df_nba = pd.read_csv("nba.csv")
```

```python
import pandas as pd
df_nba['Team'].value_counts()
```

```python
# ### 포지션별 평균 연봉은 어떻게 되나요

df_nba.groupby('Position')['Salary'].mean()
```

```python
### 팀별 평균 연봉은 어떻게 되나요?

df_nba.groupby('Team')['Salary'].mean()
```

```python
### 학교별로 출신 선수가 몇 명 인지를 보고자 하면 어떻게 하면 될까요?

df_nba['College'].value_counts().to_frame().head(10)
```

```python
 ### 연봉이 가장 많은 학교 top 10
df_nba.groupby('College')['Salary'].mean().sort_values(ascending=True).head(10)
```
